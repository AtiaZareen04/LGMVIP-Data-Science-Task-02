**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

**Read Image**

```python
img = cv2.imread('/content/img.jpg')
```

**Showing the Image using OpenCV**

```python
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Display Image using Matplotlib**

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

**Convert BGR image to RGB image**

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

**Convert Image to Pencil Sketch**




**Step1: Convert to Gray Image**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Step 2: Invert Image**

```python
invert_img = cv2.bitwise_not(gray_img)
```

**Step 3: Blur Image**

```python
blur_img = cv2.GaussianBlur(invert_img, (111, 111), 0)
```

**Step 4: Invert Blurred Image**

```python
invert_blur_img = cv2.bitwise_not(blur_img)
```

**Step 5: Sketch**

```python
sketch_img = cv2.divide(gray_img, invert_blur_img, scale = 200.0)
```

**Step 6: Save sketch**

```python
cv2.imwrite('sktech.png', sketch_img)
```

**Step 7: Display Sketch**

```python
cv2_imshow(sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
